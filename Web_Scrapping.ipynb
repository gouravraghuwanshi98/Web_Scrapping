```python
import bs4
import requests
import json as js
```

```python
link = 'https://coinmarketcap.com/historical/20210112'

soup = bs4.BeautifulSoup(requests.get(link).content,'html.parser') 
soup = soup.find('script', id="__NEXT_DATA__",type="application/json")
top100coinname = []
top100coinid = []
soup = js.loads(soup.contents[0])
soup = soup['props']['initialState']['cryptocurrency']['listingHistorical']['data']
top100coinname = []
top100coinid = []
for i in range(200):
  top100coinname.append(soup[i]['name'])
  top100coinid.append(soup[i]['id'])      

top100coinswithid=list(zip(top100coinid,top100coinname))
```
